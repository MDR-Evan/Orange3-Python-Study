## **[3] 모델 생성하기**

3.1 k-평균 알고리즘으로 데이터 학습하기

```python
X = df[['위도', '경도']]
print(X)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=6)
y_km = km.fit_predict(X[['위도', '경도']])
print(X[:10])
print(y_km[:10])
```

3.2 k-평균 군집화 결과 시각화하기

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[y_km==0]['위도'], X[y_km==0]['경도'])
plt.scatter(X[y_km==1]['위도'], X[y_km==1]['경도'])
plt.scatter(X[y_km==2]['위도'], X[y_km==2]['경도'])
plt.scatter(X[y_km==3]['위도'], X[y_km==3]['경도'])
plt.scatter(X[y_km==4]['위도'], X[y_km==4]['경도'])
plt.scatter(X[y_km==5]['위도'], X[y_km==5]['경도'])
plt.xlabel('위도')
plt.ylabel('경도')

plt.show()
```

```python
latitudeT, longitudeT = '위도', '경도'

mm = folium.Map(location=(36.3504, 127.3845), zoom_start=6)
colors = ['red', 'blue', 'green', 'purple', 'orange', 'pink']

for i in range(6):
    latitudeT =  X[y_km==i]['위도']
    longitudeT = X[y_km==i]['경도']

    for latitude, longitude in zip(latitudeT, longitudeT) :
        folium.Circle(
            location=[latitude, longitude],
            radius=10,
            color=colors[i]
        ).add_to(mm)

mm
```

## **[4] 모델 평가하기**

```python
from sklearn.metrics import silhouette_score

scores = []

for i in range(2, 8):
    km = KMeans(n_clusters=i)
    km.fit(X[['위도', '경도']])
    score = silhouette_score(X[['위도', '경도']], km.labels_)
    scores.append(score)

plt.plot(range(2, 8), scores, marker='o')
plt.xlabel('Number of clusters')
plt.show()
```
